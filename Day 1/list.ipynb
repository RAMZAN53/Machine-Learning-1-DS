# 5. Check if a List is Palindromic
## Problem Statement:
Write a function to check if a list is the same forward and backward.

## Example 1:
**Input:** `[1, 2, 3, 2, 1]`  
**Output:** `True`

## Example 2:
**Input:** `[1, 2, 3, 4, 5]`  
**Output:** `False`


```python
def Palindrome(list):
  return list==list[::-1]

list=[1, 2, 3, 2, 1]
print(Palindrome(list))
```

```python
def Palindrome(list):
    left = 0
    right = len(list) - 1
    while left < right:
        if list[left] != list[right]:
            return False
        left += 1
        right -= 1
    return True

list = [1, 2, 3, 2, 1]
print(Palindrome(list))
```

# 8. Merge Two Sorted Lists
## Problem Statement:
Write a function to merge two sorted lists into a single sorted list.

## Example 1:
**Input:** `[1, 3, 5]`, `[2, 4, 6]`  
**Output:** `[1, 2, 3, 4, 5, 6]`

## Example 2:
**Input:** `[0, 2, 8]`, `[1, 3, 7]`  
**Output:** `[0, 1, 2, 3, 7, 8]`


```python
def MergeList(lst1,lst2):
  list=lst1+lst2
  list.sort(reverse=False)
  return list

lst1=[1, 3, 5]
lst2=[2, 4, 6]
print(MergeList(lst1,lst2))
```

```python
def MergeList(lst1, lst2):
    merged_list = lst1 + lst2

    n = len(merged_list)
    for i in range(n):
        for j in range(0, n-i-1):
            if merged_list[j] > merged_list[j+1]:
                merged_list[j], merged_list[j+1] = merged_list[j+1], merged_list[j]

    return merged_list

lst1 = [1, 3, 5]
lst2 = [2, 4, 6]
print(MergeList(lst1, lst2))
```

# 9. Remove Duplicates from a List
## Problem Statement:
Write a function to remove duplicate elements from a list while maintaining order.

## Example 1:
**Input:** `[1, 2, 2, 3, 4, 4, 5]`  
**Output:** `[1, 2, 3, 4, 5]`

## Example 2:
**Input:** `["a", "b", "a", "c", "c", "d"]`  
**Output:** `["a", "b", "c", "d"]`


```python
def RemoveDuplicate(lst):
  unique= []
  seen = set()

  for item in lst:
    if item not in seen:
      unique.append(item)
      seen.add(item)

  return unique

lst = [1, 2, 2, 3, 4, 4, 5]
print(RemoveDuplicate(lst))
```

# 14. Find the Mode of a List
## Problem Statement:
Write a function to return the most frequently occurring element in a list.

## Example 1:
**Input:** `[1, 2, 2, 3, 3, 3]`  
**Output:** `3`

## Example 2:
**Input:** `["apple", "banana", "apple", "cherry", "apple"]`  
**Output:** `"apple"`


```python
def most_frequent_element(lst):
    freq_dict = {}

    for item in lst:
        if item in freq_dict:
            freq_dict[item] += 1
        else:
            freq_dict[item] = 1

    most_frequent = max(freq_dict, key=freq_dict.get)

    return most_frequent

print(most_frequent_element(["apple", "banana", "apple", "cherry", "apple"]))
```

# 15. Flatten a Nested List
## Problem Statement:
Write a function to flatten a nested list into a single list.

## Example 1:
**Input:** `[[1, 2, [3]], 4]`  
**Output:** `[1, 2, 3, 4]`

## Example 2:
**Input:** `[["a", ["b", "c"]], "d"]`  
**Output:** `["a", "b", "c", "d"]`


```python
def Flatten(lst):
    result = []
    for item in lst:
        if type(item) == list:
            result.extend(item)
        else:
            result.append(item)
    return result

print(Flatten([["a", ["b", "c"]], "d"]))
```

# 16. Find Pairs in a List That Sum to a Target Value
## Problem Statement:
Write a function to find all pairs in a list that sum to a given target.

## Example 1:
**Input:** `[1, 2, 3, 4, 5]`, `target=6`  
**Output:** `[(1, 5), (2, 4)]`

## Example 2:
**Input:** `[2, 4, 6, 8]`, `target=10`  
**Output:** `[(2, 8), (4, 6)]`


```python
def FindPairs(lst, target):
    pairs = []
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] + lst[j] == target:
                pairs.append((lst[i], lst[j]))

    return pairs

print(FindPairs([1, 2, 3, 4, 5], 6))
```

# 17. Find the Product of All Elements Except Self
## Problem Statement:
Write a function that returns a list where each index contains the product of all numbers except the number at that index.

**Output:** `[24, 12, 8, 6]`

## Example 2:
**Input:** `[5, 6, 7]`  
**Output:** `[42, 35, 30]`


```python
def Product(lst):
    total_product = 1
    for num in lst:
        total_product *= num

    result = [total_product // num for num in lst]

    return result

print(Product([1, 2, 3, 4]))
```

# 18. Find the Longest Consecutive Sequence in a List
## Problem Statement:
Write a function to find the longest sequence of consecutive numbers in a list.

## Example 1:
**Input:** `[100, 4, 200, 1, 3, 2]`  
**Output:** `4  (sequence: [1, 2, 3, 4])`

## Example 2:
**Input:** `[10, 11, 12, 50, 51, 52]`  
**Output:** `3  (sequence: [10, 11, 12])`


```python
def longest_consecutive_sequence(nums):
    num_set = set(nums)
    longest = 0

    for num in num_set:
        if num - 1 not in num_set:
            current_num = num
            current_streak = 1

            while current_num + 1 in num_set:
                current_num += 1
                current_streak += 1

            longest = max(longest, current_streak)

    return longest

print(longest_consecutive_sequence([100, 4, 200, 1, 3, 2]))
```

# 19. Check if a List Contains a Sublist
## Problem Statement:
Write a function to check if a list contains a given sublist.

## Example 1:
**Input:** `[1, 2, 3, 4, 5]`, `[3, 4]`  
**Output:** `True`

## Example 2:
**Input:** `[10, 20, 30, 40]`, `[20, 50]`  
**Output:** `False`


```python
def contains_sublist(main_list, sublist):
    if len(sublist) > len(main_list):
        return False

    for i in range(len(main_list) - len(sublist) + 1):
        if main_list[i:i + len(sublist)] == sublist:
            return True

    return False

print(contains_sublist([1, 2, 3, 4, 5], [3, 4]))
```

# 20. Find the Kth Largest Element in a List
## Problem Statement:
Write a function to find the `k`th largest element in a list.

## Example 1:
**Input:** `[3, 2, 1, 5, 6, 4]`, `k=2`  
**Output:** `5`

## Example 2:
**Input:** `[10, 20, 30, 40, 50]`, `k=3`  
**Output:** `30`


```python
import heapq

def find_kth_largest(nums, k):
    return heapq.nlargest(k, nums)[-1]

print(find_kth_largest([3, 2, 1, 5, 6, 4], 2))
```

```python
def find_kth_largest(nums, k):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] < nums[j]:
                nums[i], nums[j] = nums[j], nums[i]

    return nums[k - 1]

print(find_kth_largest([3, 2, 1, 5, 6, 4], 2))
```

# 22. Partition a List into Even and Odd Elements
## Problem Statement:
Write a function to split a list into two separate lists, one containing even numbers and the other containing odd numbers.

## Example 1:
**Input:** `[1, 2, 3, 4, 5]`  
**Output:** `([2, 4], [1, 3, 5])`

## Example 2:
**Input:** `[10, 15, 20, 25]`  
**Output:** `([10, 20], [15, 25])`


```python
def split_even_odd(nums):
    even_nums = []
    odd_nums = []

    for num in nums:
        if num % 2 == 0:
            even_nums.append(num)
        else:
            odd_nums.append(num)

    return (even_nums, odd_nums)

print(split_even_odd([1, 2, 3, 4, 5]))
```

# 23. Find the First Non-Repeating Element in a List
## Problem Statement:
Write a function to find the first element that does not repeat.

## Example 1:
**Input:** `[4, 5, 1, 2, 0, 4]`  
**Output:** `5`

## Example 2:
**Input:** `[10, 20, 10, 30, 20]`  
**Output:** `30`


```python
from collections import Counter

def first_non_repeating(nums):
    counts = Counter(nums)

    for num in nums:
        if counts[num] == 1:
            return num

    return None

print(first_non_repeating([4, 5, 1, 2, 0, 4]))
```

# 24. Sort a List Without Using Built-in Sort
## Problem Statement:
Write a function to sort a list in ascending order without using Pythonâ€™s built-in sort.

## Example 1:
**Input:** `[3, 1, 4, 1, 5, 9]`  
**Output:** `[1, 1, 3, 4, 5, 9]`

## Example 2:
**Input:** `[10, 5, 2, 8]`  
**Output:** `[2, 5, 8, 10]`


```python
def Sort(lst):
    n = len(lst)
    for i in range(n):

        for j in range(0, n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]

    return lst

print(Sort([3, 1, 4, 1, 5, 9]))
```

# 25. Find All Triplets That Sum to Zero
## Problem Statement:
Write a function to find all triplets in a list that sum to zero.

## Example 1:
**Input:** `[-1, 0, 1, 2, -1, -4]`  
**Output:** `[[-1, -1, 2], [-1, 0, 1]]`

## Example 2:
**Input:** `[1, -2, 1, 0, 5]`  
**Output:** `[[-2, 1, 1]]`


```python
def find_triplets(nums):
    nums.sort()
    result = []

    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left = i + 1
        right = len(nums) - 1

        while left < right:
            triplet_sum = nums[i] + nums[left] + nums[right]

            if triplet_sum == 0:
                result.append([nums[i], nums[left], nums[right]])
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1
            elif triplet_sum < 0:
                left += 1
            else:
                right -= 1

    return result

print(find_triplets([-1, 0, 1, 2, -1, -4]))
```
