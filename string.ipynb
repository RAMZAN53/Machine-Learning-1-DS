# 5. Find the First Non-Repeating Character in a String
## Problem Statement:
Write a function to find the first non-repeating character in a given string.

## Example 1:
**Input:** `"swiss"`  
**Output:** `"w"`

## Example 2:
**Input:** `"racecar"`  
**Output:** `"e"`


```python
def Non_Repeating(s):
    for i in range(len(s)):
        is_unique = True
        for j in range(len(s)):
            if i != j and s[i] == s[j]:
                is_unique = False
                break
        if is_unique:
            return s[i]
    return None
print(Non_Repeating("racecar"))
```

# 6. Check if Two Strings are Anagrams
## Problem Statement:
Write a function to check if two strings are anagrams.

## Example 1:
**Input:** `"listen", "silent"`  
**Output:** `True`

## Example 2:
**Input:** `"hello", "world"`  
**Output:** `False`


```python
def are_anagrams(str1, str2):
    if len(str1) != len(str2):
        return False

    str2_chars = []
    for char in str2:
        str2_chars.append(char)

    for char1 in str1:
        found = False
        for i in range(len(str2_chars)):
            if char1 == str2_chars[i]:
                str2_chars[i] = None
                found = True
                break
        if not found:
            return False

    return True

print(are_anagrams("listen", "silent"))
```

# 7. Convert a String to Uppercase
## Problem Statement:
Write a function to convert a string to uppercase.

## Example 1:
**Input:** `"python"`  
**Output:** `"PYTHON"`

## Example 2:
**Input:** `"Hello World"`  
**Output:** `"HELLO WORLD"`


```python
def to_uppercase(s):
    result = ""
    for char in s:
        if 'a' <= char <= 'z':
            upper_char = chr(ord(char) - 32)
            result += upper_char
        else:
            result += char
    return result
```

# 8. Find the Most Frequent Character in a String
## Problem Statement:
Write a function to find the most frequently occurring character in a string.

## Example 1:
**Input:** `"banana"`  
**Output:** `"a"`

## Example 2:
**Input:** `"mississippi"`  
**Output:** `"i"`


```python
def most_frequent_char(s):
    max_count = 0
    frequent_char = None

    for i in range(len(s)):
        count = 0
        for j in range(len(s)):
            if s[i] == s[j]:
                count += 1

        if count > max_count:
            max_count = count
            frequent_char = s[i]

    return frequent_char

print(most_frequent_char("banana"))
```

# 10. Find the Length of the Longest Word in a Sentence
## Problem Statement:
Write a function to find the length of the longest word in a sentence.

## Example 1:
**Input:** `"The quick brown fox"`  
**Output:** `5`

## Example 2:
**Input:** `"A journey of a thousand miles"`  
**Output:** `8`


```python
def Longest_Word(sentence):
    length = 0
    max_length = 0

    for i in range(len(sentence)):
        if sentence[i] != ' ':
            length += 1
        else:
            if length > max_length:
                max_length = length
            length = 0

    if length > max_length:
        max_length = length

    return max_length

print(Longest_Word("The quick brown fox"))
```

# 13. Reverse Each Word in a Sentence
## Problem Statement:
Write a function to reverse each word in a sentence while maintaining word order.

## Example 1:
**Input:** `"hello world"`  
**Output:** `"olleh dlrow"`

## Example 2:
**Input:** `"Python is fun"`  
**Output:** `"nohtyP si nuf"`


```python
def reverse_each_word(sentence):
    words = sentence.split()
    result = []

    for word in words:
        reversed_word = ""
        i = len(word) - 1
        while i >= 0:
            reversed_word += word[i]
            i -= 1
        result.append(reversed_word)

    final_sentence = ""
    for i in range(len(result)):
        final_sentence += result[i]
        if i != len(result) - 1:
            final_sentence += " "

    return final_sentence

reverse_each_word("Python is fun")
```

# 15. Check if a String is a Rotation of Another String
## Problem Statement:
Write a function to check if one string is a rotation of another.

## Example 1:
**Input:** `"waterbottle", "erbottlewat"`  
**Output:** `True`

## Example 2:
**Input:** `"hello", "lohel"`  
**Output:** `True`


```python
def is_rotation(s1, s2):
    if len(s1) != len(s2):
        return False

    combined = s1 + s1

    if s2 in combined:
        return True
    return False

print(is_rotation("waterbottle", "erbottlewat"))
print(is_rotation("hello", "lohel"))
```

# 16. Find the Second Most Frequent Character in a String
## Problem Statement:
Write a function to find the second most frequent character in a string.

## Example 1:
**Input:** `"banana"`  
**Output:** `"n"`

## Example 2:
**Input:** `"mississippi"`  
**Output:** `"s"`


```python
def second_most_frequent(s):
    freq_dict = {}

    for ch in s:
        if ch in freq_dict:
            freq_dict[ch] += 1
        else:
            freq_dict[ch] = 1

    sorted_freq = sorted(freq_dict.items(), key=lambda x: x[1], reverse=True)
    return sorted_freq[1][0]

print(second_most_frequent("banana"))
print(second_most_frequent("mississippi"))
```

# 19. Remove Duplicate Characters from a String

Write a function to remove duplicate characters from a string.

## Example 1:
**Input:** `"hello"`  
**Output:** `"helo"`

## Example 2:
**Input:** `"mississippi"`  
**Output:** `"misp"`


```python
def remove_duplicates(s):
    seen = []
    result = ""
    for char in s:
        if char not in seen:
            result += char
            seen.append(char)

    return result

print(remove_duplicates("hello"))
```

# 20. Sort a String Alphabetically
## Problem Statement:
Write a function to sort the characters of a string alphabetically.

## Example 1:
**Input:** `"hello"`  
**Output:** `"ehllo"`

## Example 2:
**Input:** `"python"`  
**Output:** `"hnopty"`


```python
def sort_string(s):
    char_list = list(s)

    n = len(char_list)
    for i in range(n):
        for j in range(0, n-i-1):
            if char_list[j] > char_list[j+1]:
                char_list[j], char_list[j+1] = char_list[j+1], char_list[j]

    return ''.join(char_list)

print(sort_string("hello"))
```

# 21. Find the Smallest Window in a String Containing All Characters of Another String
## Problem Statement:
Write a function to find the smallest window in a string that contains all characters of another string.

## Example 1:
**Input:** `"ADOBECODEBANC", "ABC"`  
**Output:** `"BANC"`

## Example 2:
**Input:** `"aabbcc", "abc"`  
**Output:** `"abc"`


```python
def smallest_window(s1, s2):
    len_s2 = len(s2)
    substrings = []

    for i in range(len(s1) - len_s2 + 1):
        substrings.append(s1[i:i+len_s2])

    def contains_all_chars(sub, target):
        target_dict = {}
        for char in target:
            if char in target_dict:
                target_dict[char] += 1
            else:
                target_dict[char] = 1

        for char in sub:
            if char in target_dict:
                target_dict[char] -= 1
            if target_dict[char] == 0:
                del target_dict[char]

        return len(target_dict) == 0

    smallest_window = ""
    for window in substrings:
        if contains_all_chars(window, s2):
            if not smallest_window or len(window) < len(smallest_window):
                smallest_window = window

    return smallest_window

print(smallest_window("ADOBECODEBANC", "ABC"))
```

```python
from collections import Counter
def contains_all_chars(substring, target):
    substring_count = Counter(substring)
    target_count = Counter(target)
    for char, count in target_count.items():
        if substring_count[char] < count:
            return False
    return True
def Split(lst):
    substrings = []
    for i in range(0, len(lst), 3):
        substrings.append(lst[i:i+3])
    return substrings
def Matching_Substrings(s, target):
    substrings = Split(s)
    matching_substrings = []

    for substring in substrings:
        if contains_all_chars(substring, target):
            matching_substrings.append(substring)

    return matching_substrings

print(Matching_Substrings("ADOBECODEBANC", "ABC"))
```

# 23. Find All Permutations of a String
## Problem Statement:
Write a function to generate all permutations of a given string.

## Example 1:
**Input:** `"abc"`  
**Output:** `["abc", "acb", "bac", "bca", "cab", "cba"]`

## Example 2:
**Input:** `"ab"`  
**Output:** `["ab", "ba"]`


```python
def generate_permutations(s):
    if len(s) == 1:
        return [s]

    permutations = []

    for i in range(len(s)):
        current_char = s[i]
        remaining_str = s[:i] + s[i+1:]

        for perm in generate_permutations(remaining_str):
            permutations.append(current_char + perm)

    return permutations

print(generate_permutations("abc"))
```
